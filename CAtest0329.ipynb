```python
import pandas as pd
import matplotlib.pyplot as plt
from rpy2 import robjects as ro
from rpy2.robjects import pandas2ri
from rpy2.robjects.conversion import localconverter

# 激活Pandas DataFrame到R data.frame的自动转换
pandas2ri.activate()

# 加载INLA库
ro.r('library(INLA)')

def run_inla_model(formula, data, model_name):
    with localconverter(ro.default_converter + pandas2ri.converter):
        r_df = ro.conversion.py2rpy(data)
        ro.r.assign("r_df", r_df)  # 将r_df传递到R环境

    # 在R中创建滞后变量
    if 'lag' in formula:
        formula_parts = formula.split('~')
        lags = []
        for term in formula_parts[1].split('+'):
            if 'lag' in term:
                lag_term = term.strip().split('_')[-1]  # e.g., 'lag1'
                lag = int(''.join(filter(str.isdigit, lag_term)))  # 提取数字部分
                lags.append(lag)
        for lag in lags:
            ro.r(f'r_df$HeatCount_lag{lag} <- c({",".join(["NA"]*lag)}, head(r_df$HeatCount, -{lag}))')

    inla_call = f"""
    result <- inla(formula = {formula}, data = r_df, family = 'gaussian',
                   control.predictor = list(compute = TRUE))
    """
    ro.r(inla_call)
    ro.r(f'print(summary(result), digits = 3)')
    mlik = ro.r('result$mlik')[0]
    print(f"模型: {model_name}")
    print(f"边际对数似然: {mlik}")
    print()
    return mlik

# 加载数据
df = pd.read_csv('/content/TestData_California.csv')

# 存储模型边际对数似然值
mliks = []
models = []

# 运行不包含滞后变量的模型
base_formula = "SentimentScore ~ HeatCount"
print("运行不包含滞后变量的模型")
base_mlik = run_inla_model(base_formula, df, "基础模型")
mliks.append(base_mlik)
models.append("基础模型")

# 运行包含不同滞后天数的模型
for lag_days in range(1, 8):  # 滞后1到7天
    lagged_formula = "SentimentScore ~ HeatCount"
    for i in range(1, lag_days + 1):
        lagged_formula += f" + HeatCount_lag{i}"  # 每天滞后
    model_name = f"滞后{lag_days}天模型"
    print(f"运行包含滞后{lag_days}天的模型")
    mlik = run_inla_model(lagged_formula, df, model_name)
    mliks.append(mlik)
    models.append(model_name)

# 绘制边际对数似然值变化趋势
plt.figure(figsize=(12, 6))
plt.plot(models, mliks, marker='o')
plt.xticks(rotation=45)
plt.ylabel('边际对数似然值')
plt.title('模型边际对数似然值变化趋势')
plt.show()
```

要确定模型结果是否表现出空间自相关性，可以使用 Moran's I 统计量来衡量。Moran's I 是一种评估空间自相关的方法，其值范围通常在 -1（完全负相关）到 +1（完全正相关）之间，值接近 0 表示随机分布（无空间自相关）。

```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Polygon
from rpy2 import robjects as ro
from rpy2.robjects import pandas2ri
from rpy2.robjects.conversion import localconverter
from libpysal.weights import Queen
from esda.moran import Moran
import matplotlib.pyplot as plt

# 创建虚拟地理数据
county_names = [
    'Alameda', 'Alpine', 'Amador', 'Butte', 'Calaveras', 'Colusa',
    'Contra Costa', 'Del Norte', 'El Dorado', 'Fresno', 'Glenn',
    'Humboldt', 'Imperial', 'Inyo', 'Kern', 'Kings', 'Lake', 'Lassen',
    'Los Angeles', 'Madera', 'Marin', 'Mariposa', 'Mendocino', 'Merced',
    'Modoc', 'Mono', 'Monterey', 'Napa', 'Nevada', 'Orange', 'Placer',
    'Plumas', 'Riverside', 'Sacramento', 'San Benito', 'San Bernardino',
    'San Diego', 'San Francisco', 'San Joaquin', 'San Luis Obispo', 'San Mateo',
    'Santa Barbara', 'Santa Clara', 'Santa Cruz', 'Shasta', 'Sierra',
    'Siskiyou', 'Solano', 'Sonoma', 'Stanislaus', 'Sutter', 'Tehama',
    'Trinity', 'Tulare', 'Tuolumne', 'Ventura', 'Yolo', 'Yuba'
]
county_polygons = [Polygon([(i, 0), (i+1, 0), (i+1, 1), (i, 1)]) for i in range(len(county_names))]
county_data = gpd.GeoDataFrame({'geometry': county_polygons, 'CountyName': county_names})

# 生成随机空间邻接矩阵
np.random.seed(42)
adjacency_matrix = np.random.randint(0, 2, size=(len(county_names), len(county_names)))
np.fill_diagonal(adjacency_matrix, 0)
adjacency_df = pd.DataFrame(adjacency_matrix, index=county_names, columns=county_names)
r_adjacency = pandas2ri.py2rpy(adjacency_df)

# 激活Pandas DataFrame到R data.frame的自动转换
pandas2ri.activate()

# 在R中安装并加载所需的包
ro.r('''
if (!requireNamespace("INLA", quietly = TRUE)) {
    install.packages("INLA", repos = "https://inla.r-inla-download.org/R/stable", type = "source")
}
library(INLA)
if (!requireNamespace("spdep", quietly = TRUE)) {
    install.packages("spdep")
}
library(spdep)
''')

def run_inla_model(formula, data, model_name, adjacency=None):
    with localconverter(ro.default_converter + pandas2ri.converter):
        r_df = ro.conversion.py2rpy(data)
        ro.r.assign("r_df", r_df)

        if adjacency is not None:
            ro.r.assign("adjacency", adjacency)

        inla_call = f"""
        result <- inla(formula = {formula}, data = r_df, family = 'gaussian',
                       control.predictor = list(compute = TRUE),
                       control.family = list(hyper = list(prec = list(initial = 25, fixed = FALSE))),
                       control.inla = list(strategy = 'adaptive'),
                       control.compute = list(dic = TRUE, waic = TRUE),
                       E = adjacency)
        """
        ro.r(inla_call)
        ro.r('print(summary(result), digits = 3)')

        mlik = ro.r('result$mlik')[0]
        return mlik

# 读取数据
df = pd.read_csv('/content/TestData_California.csv')

# 存储模型边际对数似然值
county_results = {}

# 按县分组并运行模型
for county_name in county_names:
    group_data = df[df['CountyName'] == county_name].copy()
    if group_data.empty:
        continue

    group_data.sort_values('Date', inplace=True)

    # 创建滞后变量
    for lag in range(1, 15):
        group_data[f'HeatCount_lag{lag}'] = group_data['HeatCount'].shift(lag)
    group_data.dropna(inplace=True)

    lagged_formula = "SentimentScore ~ " + " + ".join([f"HeatCount_lag{lag}" for lag in range(1, 15)])

    model_name = f"{county_name}_14_day_lag"
    mlik = run_inla_model(lagged_formula, group_data, model_name, adjacency=r_adjacency)
    county_results[county_name] = mlik

# 将结果加入GeoDataFrame
county_data['MLIK'] = county_data['CountyName'].map(county_results)

# 创建空间权重矩阵
w = Queen.from_dataframe(county_data)

# 计算Moran's I
mi_result = Moran(county_data['MLIK'].dropna(), w)

# 输出Moran's I统计量
print(f"Moran's I for Marginal Log Likelihood: {mi_result.I}, p-value: {mi_result.p_sim}")

# 可视化
fig, ax = plt.subplots(figsize=(12, 10))
county_data.plot(column='MLIK', cmap='coolwarm', legend=True, ax=ax)
ax.set_title("Marginal Log Likelihood with Spatial Autocorrelation", fontsize=15)
plt.show()
```
